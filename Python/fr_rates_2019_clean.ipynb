```python
cc = rates.nom.unique() #create a list of all communities of cities 
```

```python
print(type(cc))
```

```python
map = pd.read_csv('nature_cat.csv')
```

```python
map_dict = map.to_dict('index')
```

```python
cc_dict = {}
for i in cc:
    cc_dict[i] = map_dict
next(iter(cc_dict))
```

```python
cc_dict['CC RIVES DE L\'AIN']
```

```python
print(type(cc_dict['CC RIVES DE L\'AIN']))
```

```python
len(cc_dict.keys())
```

```python
df_cc_cat = pd.DataFrame.from_dict({(i,j): cc_dict[i][j] 
                           for i in cc_dict.keys() 
                           for j in cc_dict[i].keys()},
                       orient='index')
print(df_cc_cat.head())
```

```python
df_cc_cat.reset_index(level=0, inplace = True)
```

```python
print(df_cc_cat.tail())
```

```python
df_cc_cat.columns
```

```python
df_cc_cat.index.values
```

```python
df_cc_cat.columns = ['nom', 'nat_cat_id', 'natureID', 'natureName', 'categoryID',
       'categoryName']
```

```python
nat = rates[['nom','natureId','regime']]
nat = nat.dropna()
```

```python
df_nat_rates = pd.merge(df_cc_cat, nat, how='left', left_on=['natureID','nom'], right_on = ['natureId','nom'])
```

```python
print(df_nat_rates.head())
```

```python
cat = rates[['nom','dateDebut','dateFin','tarif','categorieId']]
cat = cat.dropna()
```

```python
df_nat_cat_rates= pd.merge(df_nat_rates, cat,  how='outer', left_on=['categoryID','nom'], right_on = ['categorieId','nom'])
```

```python
df_nat_cat_rates.head(200)
```

```python
df_nat_cat_rates.drop(columns=['natureId','categorieId'])
df_nat_cat_rates = df_nat_cat_rates.dropna()
```

```python
df_nat_cat_rates.to_csv('fr_cc_rates.csv', sep='|', index = False, encoding='utf-8')
```

```python
city = rates[['siren','nom','numero','libelle5','loyerMaximumNuit','loyerMaximumSemaine','loyerMaximumMois','taxeAdditionnelleDepartementale','siren7','nom8']]
```

```python
df_final= pd.merge(df_nat_cat_rates, city,  how='outer', left_on=['nom'], right_on = ['nom'])
```

```python
df_final = df_final.dropna()
```
